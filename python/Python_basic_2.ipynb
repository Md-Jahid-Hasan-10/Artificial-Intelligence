## 1. Tuple
A tuple is a collection of items that is ordered and immutable (cannot be changed).

```python
# Creating a tuple
numbers = (1, 2, 3, 4)
names = ("Jahid", "Ali", "Sara")

print(numbers)
print(names)
```

### Tuple Properties
- Ordered
- Immutable (cannot modify)
- Allows duplicate values

```python
# Access tuple elements
numbers = (10, 20, 30)

print(numbers[0])
print(numbers[1])
```

## 2. Set
A set is an unordered collection of unique elements (no duplicates).

```python
# Creating a set
numbers = {1, 2, 3, 4}
duplicates = {1, 2, 2, 3}

print(numbers)
print(duplicates)  # duplicates removed automatically
```

### Set Operations
- add()
- remove()
- union()
- intersection()

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))         # combine
print(a.intersection(b))  # common elements
```
