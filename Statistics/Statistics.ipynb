## 1. What is Statistics?
Statistics is the study of collecting, analyzing, and interpreting data. It is very important in Machine Learning.

```python
# Example dataset
data = [10, 20, 30, 40, 50]
print(data)
```

## 2. Mean (Average)
Mean is the average of all values.
Formula: sum of values / total number of values

```python
data = [10, 20, 30, 40, 50]

mean = sum(data) / len(data)
print("Mean:", mean)
```

## 3. Median
Median is the middle value after sorting the data.

```python
import statistics

data = [10, 20, 30, 40, 50]
print("Median:", statistics.median(data))
```

## 4. Mode
Mode is the value that appears most frequently.

```python
import statistics

data = [1, 2, 2, 3, 4]
print("Mode:", statistics.mode(data))
```

## 5. Variance
Variance measures how far data points are spread from the mean.

```python
import statistics

data = [10, 20, 30, 40, 50]
print("Variance:", statistics.variance(data))
```

## 6. Standard Deviation
Standard deviation is the square root of variance.

```python
import statistics

data = [10, 20, 30, 40, 50]
print("Standard Deviation:", statistics.stdev(data))
```

## 7. Min, Max, Range
- Min = smallest value
- Max = largest value
- Range = max - min

```python
data = [10, 20, 30, 40, 50]

print("Min:", min(data))
print("Max:", max(data))
print("Range:", max(data) - min(data))
```

## 8. Introduction to NumPy
NumPy is used for numerical operations in Python.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
```

## What is Correlation?

Correlation measures the strength and direction of the relationship between two variables.

- Range: -1 to +1
- +1 → Strong positive correlation
- -1 → Strong negative correlation
- 0 → No correlation

## Correlation Formula

r = Σ((x - mean(x)) (y - mean(y))) / sqrt(Σ(x - mean(x))² * Σ(y - mean(y))²)

```python
hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]
correlation = np.corrcoef(hours, marks)[0,1]
print("Correlation:", correlation)
```

## Interpretation

If correlation is close to:
- 1 → strong positive relationship
- 0 → no relationship
- -1 → strong negative relationship

## What is Linear Regression?

Linear Regression is used to predict a dependent variable based on an independent variable.

It finds the best-fit straight line.

## Linear Equation

y = mx + b

Where:
- m = slope
- b = intercept

## Real-Life Example: Salary Prediction

Predict salary based on years of experience.

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([30000, 35000, 40000, 45000, 50000])
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[6]])
print("Predicted Salary for 6 years experience:", prediction[0])
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Linear Regression Example")
plt.show()
```

## Conclusion

- Correlation measures relationship between variables
- Linear Regression predicts future values
- Both are important in Machine Learning and Data Analysis

```python
model.score(X,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## What is Polynomial Regression?

Polynomial Regression is used when the relationship between variables is NOT linear.

Instead of a straight line, it fits a curve.

Example:
- Linear → straight line
- Polynomial → curved line

## Polynomial Equation

y = b0 + b1x + b2x² + b3x³ + ... + bnxⁿ

## Example Dataset (Non-Linear)

Study hours vs performance (non-linear pattern)

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 9, 16, 25])   # curve (square relationship)
```

## Linear Regression (Wrong Fit)

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
```

```python
prediction = lin_model.predict([[6]])
print("Predicted Salary for 6 years experience:", prediction[0])
```

```python
print("Linear Score")
lin_model.score(X, y)
```

## Convert to Polynomial Features

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

## Train Polynomial Regression Model

```python
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

## Prediction

```python
pred = poly_model.predict(poly.transform([[6]]))
print("Prediction for x=6:", pred[0])
```

## Visualization (Linear vs Polynomial)

```python
plt.scatter(X, y)

# Linear line
plt.plot(X, lin_model.predict(X), linestyle='dashed')

# Polynomial curve
plt.plot(X, poly_model.predict(X_poly))

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial vs Linear Regression")
plt.show()
```

## Conclusion

- Linear Regression → straight line
- Polynomial Regression → curve
- Use Polynomial when data is non-linear
- Degree controls complexity of curve

```python
print("Polynomial Score")
poly_model.score(X_poly, y)
```
