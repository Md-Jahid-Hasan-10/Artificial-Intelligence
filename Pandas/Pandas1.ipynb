```python
import os
root_dir = "G:\\ML\\Numpy"
data_dir = os.path.join(root_dir,'Python_')
dataset_path = os.path.join(data_dir,'data2.csv')
print(dataset_path)
```

```python
sales = pd.read_csv(dataset_path)
sales.head()
```

```python
sales = sales.drop(columns = ['smoker'])
sales.head()
```

```python
def normalize(col):
    mean = col.mean()
    std = col.std()
    return (col - mean) / std
```

```python
normalize(sales['weight'])
```

```python
import numpy as np
apply_two_col = sales[['height','weight']].agg([normalize,np.cumsum])
apply_two_col
```
